```python
import numpy as np
import pandas as pd
```

```python

# Chỉ định mã hóa chính xác (ví dụ: 'latin1') khi đọc tệp CSV
all_data_cleaned_3 = pd.read_csv('all_data_cleaned_3.csv')

# Hiển thị DataFrame
all_data_cleaned_3
```

```python
all_data_cleaned_3
```

```python
# Số ngày bạn muốn dự đoán
num_days_to_predict = 1

# Tạo các cột mới cho lượng mưa của ngày sau, 2 ngày sau, ..., 7 ngày sau
for i in range(1, num_days_to_predict + 1):
    col_name = f'TS_{i}_DAY_AFTER'
    all_data_cleaned_3[col_name] = all_data_cleaned_3['TS'].shift(-i)

# In ra một số dòng đầu của dataframe với các cột mới
all_data_cleaned_3[['DATE', 'TS'] + [f'TS_{i}_DAY_AFTER' for i in range(1, num_days_to_predict + 1)]].head()
```

```python
all_data_cleaned_3 = all_data_cleaned_3.dropna(subset=['TS_1_DAY_AFTER'])
```

```python
all_data_cleaned_3['DATE'] = pd.to_datetime(all_data_cleaned_3['DATE'])

# Hiển thị thông tin sau khi chuyển đổi
all_data_cleaned_3.info()
```

```python
all_data_cleaned_3
```

```python
#In dữ liệu ra file csv
all_data_cleaned_3.to_csv(r'data_2024_final.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=all_data_cleaned_3)
plt.title('Box Plot of Features in all_data_cleaned_3')
plt.xticks(rotation=90)
plt.show()

# Scatter Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=all_data_cleaned_3, x='DATE', y='TS')
plt.title('Scatter Plot of DATE vs TS')
plt.xticks(rotation=90)
plt.show()

# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=all_data_cleaned_3['TS'], bins=20, kde=True)
plt.title('Histogram of TS')
plt.show()
```
