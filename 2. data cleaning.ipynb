```python
import pandas as pd

# Tạo một biến DataFrame để lưu toàn bộ dữ liệu
all_data = pd.DataFrame()

# Tên các cột chung
common_columns = ['YEAR', 'DOY']

# Đọc dữ liệu từ các tệp CSV và loại bỏ các dòng đầu không cần thiết
df1 = pd.read_csv('data/nasa_power_data_part1.csv', skiprows=25)
df2 = pd.read_csv('data/nasa_power_data_part2.csv', skiprows=24)
df3 = pd.read_csv('data/nasa_power_data_part3.csv', skiprows=25)
df4 = pd.read_csv('data/nasa_power_data_part4.csv', skiprows=24)
df5 = pd.read_csv('data/nasa_power_data_part5.csv', skiprows=25)
df6 = pd.read_csv('data/nasa_power_data_part6.csv', skiprows=24)
df7 = pd.read_csv('data/nasa_power_data_part7.csv', skiprows=25)
df8 = pd.read_csv('data/nasa_power_data_part8.csv', skiprows=24)

# Ghép từng cặp DataFrame theo các cột chung
merged_df1 = pd.merge(df1, df2, on=common_columns)
merged_df2 = pd.merge(df3, df4, on=common_columns)
merged_df3 = pd.merge(df5, df6, on=common_columns)
merged_df4 = pd.merge(df7, df8, on=common_columns)

# Nối tất cả các DataFrame đã ghép lại vào một DataFrame duy nhất
all_data = pd.concat([merged_df1, merged_df2, merged_df3, merged_df4])

# In toàn bộ dữ liệu trên màn hình
all_data
```

```python
# Đặt ngưỡng phần trăm bị thiếu mà bạn quan tâm
threshold_percentage = 80

# Đếm số biến bị thiếu theo dòng
missing_by_row = all_data[all_data == -999.00].count(axis=1)

# Tổng số biến trong mỗi dòng
total_columns = all_data.shape[1]

# Tính phần trăm biến bị thiếu theo dòng
missing_percentage_by_row = (missing_by_row / total_columns) * 100

# Tìm các dòng có phần trăm biến bị thiếu lớn hơn ngưỡng
rows_with_high_missing = all_data[missing_percentage_by_row > threshold_percentage]

# In kết quả
rows_with_high_missing
```

```python
# Lấy chỉ mục (index) của các hàng có phần trăm biến bị thiếu lớn hơn ngưỡng
rows_to_remove = missing_percentage_by_row[missing_percentage_by_row > threshold_percentage].index

# Xóa các hàng có dữ liệu thiếu trên 80%
all_data_cleaned_1 = all_data.drop(index=rows_to_remove)
all_data_cleaned_1
```

```python
# Đếm số lượng giá trị bị thiếu (-999.00)
missing_count = (all_data_cleaned_1 == -999).sum()

# Tổng số giá trị
total_count = all_data_cleaned_1.shape[0]

# Phần trăm giá trị bị thiếu
missing_percentage = (missing_count / total_count) * 100

# Tạo DataFrame kết quả
missing_info = pd.DataFrame({'Số lượng bị thiếu': missing_count, 'Phần trăm bị thiếu (%)': missing_percentage})

# In kết quả
missing_info
```

```python
# In ra các giá trị trong bảng missing_info với điều kiện số lượng bị thiếu lớn hơn 0
missing_info[missing_info['Số lượng bị thiếu'] > 0]
```

```python
# Các cột cần bỏ
columns_to_remove = [ 'ALLSKY_SFC_SW_DNI', 'ALLSKY_SFC_SW_DIFF', 'CLRSKY_SFC_PAR_TOT', 'ALLSKY_SFC_UVB', 'ALLSKY_SFC_UVA', 'ALLSKY_SFC_UV_INDEX']

# Xóa các cột sau
all_data_cleaned_2 = all_data_cleaned_1.drop(columns=columns_to_remove)

# Danh sách các cột cần thay thế giá trị bị thiếu bằng giá trị trung bình
columns_to_fill = [
    'ALLSKY_SFC_SW_DWN', 'CLRSKY_SFC_SW_DWN', 'TOA_SW_DWN', 
    'ALLSKY_SFC_PAR_TOT', 'ALLSKY_SFC_LW_DWN', 'GWETTOP'
]

# Thay thế giá trị bị thiếu (được định nghĩa là -999) bằng giá trị trung bình của cột đó
for column in columns_to_fill:
    mean_value = all_data_cleaned_1[all_data_cleaned_1[column] != -999][column].mean()
    all_data_cleaned_2[column].replace(-999, mean_value, inplace=True)

# In toàn bộ dữ liệu đã xử lý
all_data_cleaned_2
```

```python
# Đếm số lượng giá trị bị thiếu (-999.00)
missing_count_2 = (all_data_cleaned_2 == -999).sum()

# Tổng số giá trị
total_count_2 = all_data_cleaned_2.shape[0]

# Phần trăm giá trị bị thiếu
missing_percentage_2 = (missing_count_2 / total_count_2) * 100

# Tạo DataFrame kết quả
missing_info_2 = pd.DataFrame({'Số lượng bị thiếu': missing_count_2, 'Phần trăm bị thiếu (%)': missing_percentage_2})

# In kết quả
missing_info_2
```

```python
# Chuyển cột DOY thành ngày và tháng
all_data_cleaned_2['DATE'] = pd.to_datetime(all_data_cleaned_2['YEAR'].astype(str) + all_data_cleaned_2['DOY'].astype(str), format='%Y%j')
all_data_cleaned_2['MONTH'] = all_data_cleaned_2['DATE'].dt.month
all_data_cleaned_2['DAY'] = all_data_cleaned_2['DATE'].dt.day

# In ra DataFrame sau khi xử lý
all_data_cleaned_2
```

```python
# Sao chép DataFrame hiện có
all_data_cleaned_3 = all_data_cleaned_2.copy()

# Chuyển cột DOY thành ngày và tháng
all_data_cleaned_3['DATE'] = pd.to_datetime(all_data_cleaned_3['YEAR'].astype(str) + all_data_cleaned_3['DOY'].astype(str), format='%Y%j')
all_data_cleaned_3['MONTH'] = all_data_cleaned_3['DATE'].dt.month
all_data_cleaned_3['DAY'] = all_data_cleaned_3['DATE'].dt.day

# Đặt các cột mới bên phải cột DOY
all_data_cleaned_3 = all_data_cleaned_3[['DATE','YEAR', 'MONTH', 'DAY'] + [col for col in all_data_cleaned_3.columns if col not in ['LAT', 'LON', 'YEAR', 'DOY', 'DATE', 'MONTH', 'DAY']]]

# In ra DataFrame sau khi xử lý
all_data_cleaned_3
```

Loại bỏ các outlier

```python
import numpy as np

# Lấy các cột số
numeric_cols = all_data_cleaned_3.select_dtypes(include=np.number).columns
df_numeric = all_data_cleaned_3[numeric_cols]

# Tính Z-score
z_scores = np.abs((df_numeric - df_numeric.mean()) / df_numeric.std())

# Tìm các giá trị ngoại lai
outliers = (z_scores > 4).any(axis=1)

# Loại bỏ các giá trị ngoại lai
all_data_cleaned_3 = all_data_cleaned_3[~outliers]

# Hiển thị 5 dòng đầu tiên của dataframe đã lọc
print("DataFrame sau khi loại bỏ ngoại lai:")
print(all_data_cleaned_3.head().to_markdown(index=False, numalign="left", stralign="left"))

# Hiển thị số lượng mẫu đã bị loại bỏ
num_outliers_removed = outliers.sum()
print(f"\nSố lượng mẫu đã bị loại bỏ: {num_outliers_removed}")
```
